# L'analyse des données

```python
# Import package
import numpy as np
import pandas as pd
```

## Sélectionner la donnée

```python
# Créeons une liste à 2D
data_list = [
    [ 0,  1,  2,  3,  4],
    [10, 11, 12, 13, 14],
    [20, 21, 22, 23, 24],
    [30, 31, 32, 33, 34],
    [40, 41, 42, 43, 44],
]

data_np = np.array(data_list)
data_np
```

### Les lignes (rows)

```python
# data_np[row(s), column(s)]
data_np[0] # [ 0,  1,  2,  3,  4]
data_np[:2] # [[ 0,  1,  2,  3,  4], [10, 11, 12, 13, 14]]
data_np[1:3] # [[10, 11, 12, 13, 14], [20, 21, 22, 23, 24]]
```

### Les colonnes (columns)

```python
# data_np[row(s), column(s)]
data_np[:,0] # [ 0, 10, 20, 30, 40]
data_np[2:,0] # [20, 30, 40]
data_np[2,0] # 20
```

### La sélection par étape

```python
# np_array[start:stop:step]
data_np[0][0:4:2] # [0, 2]
```

## Manipuler la donnée

### La vectorisation

```python
# La vectorisation la méthode la plus rapide et la moins gourmande en mémoire pour manipuler la donnée d'un 
# np_array ou d'un DataFrame comparé à une itération issue de 'for loop', '.apply()' ou '.map'
my_list = [
    [6, 5],
    [1, 3],
    [5, 6],
    [1, 4],
    [3, 7],
    [5, 8],
    [3, 5],
    [8, 4],
]

my_list_np = np.array(my_list)

my_sum = my_list_np[:, 0] + my_list_np[:, 1] # vectorial "+" operator
my_sum
```

### Le boolean indexing

```python
my_list_np[my_list_np == 5] = 99
my_list_np
```

### Le regroupement

```python
df = pd.DataFrame({'Student Names' : ['Jenny', 'Singh', 'Charles', 'Richard', 'Veena'],
                   'Category' : ['Online', 'Offline', 'Offline', 'Offline', 'Online'],
                   'Gender' : ['Female', 'Male', 'Male', 'Male', 'Female'],
                  'Courses': ['Java', 'Spark', 'PySpark','Hadoop','C'],
                   'Fee': [15000, 17000, 27000, 29000, 12000],
                   'Discount': [1100, 800, 1000, 1600, 600]})
df
```

#### df.groupby()

```python
# La méthode '.groupby()' permet de regrouper un DataFrame en fonction des colonnes sélectionner. Cette méthode
# s'applique toujours avec une méthode d'agrrégation
df.groupby(['Gender','Student Names']).mean() # count(), max(), min(), sum()
```

#### df.pivot_table()

```python
# La méthode '.pivot' permet de faire pivoter un tableau en fonction des valeurs des colonnes
pd.pivot(df, index='Student Names', columns="Category")
```

### Multiple files

```python
df1 = pd.DataFrame([['a', 1], ['b', 2]], columns=['letter', 'number'])
df2 = pd.DataFrame([['a', 3], ['b', 4]], columns=['letter', 'number'])
```

#### pd.concat()

```python
# Utiliser '.concat()' va créer un nouveau DataFrame avec les DFs les uns après les autres sans changer les index
pd.concat([df1,df2])
pd.concat([df1,df2], ignore_index=True) # le résultat sera le même mais le nouveau DataFrame sera re-indexé
```

#### pd.DataFrame.merge()

```python
# Le '.merge()' va se comporter comme une requête SQL JOIN, le nouveau DataFrame sera une fusion des DFs 
# sur la key sélectionnée
df1.merge(df2, on="letter", how="inner") # outer, left, right
```

#### pd.DataFrame.join()

```python
# Le '.join()' va se comporter comme le '.merge()' à la différence près que la fusion ne se fait pas sur une
# key en particulier mais sur chacune de chaque DF
df1.join(df2, lsuffix='_df1', rsuffix='_df2')
```

## Les bases

### Le plot simple

```python
years_x = [1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
total_y = [1243, 1543, 1619, 1831, 1960, 2310, 2415, 2270, 1918]
coal_y = [823, 1136, 1367, 1547, 1660, 1927, 1983, 1827, 1352]
gas_y = [171, 200, 166, 175, 228, 280, 319, 399, 529]
```

```python
# Pour qu'un plot puisse être généré il faut que les variables x et y soit de dimension égale
```

```python
with plt.style.context('seaborn'): # permet d'appliquer un style à notre plot
    # On peut déterminer la taille de la figure
    plt.figure(figsize=(10,5))
    # On assigne aux abscisses 'years_x' et 'total_y', 'coal_y'et 'gas_y' aux ordonnées
    plt.plot(years_x, total_y, label="Total", linestyle=':', marker='s') 
    plt.plot(years_x, coal_y, label="Coal")
    plt.plot(years_x, gas_y, label="Gaz")
    # On ajouter un titre
    plt.title('USA - CO2 emissions from electricity production')
    # On ajoute un label aux abscisses
    plt.xlabel("Year")
    # On ajoute un label aux ordonnées
    plt.ylabel("CO2 - M of tons")
    # Il est obligatoire si l'on souhaite ajouter des labels à chaque courbes de définir une légende
    plt.legend()
    # '.grid' permet de renforcer la lecture de notre plot en ajoutant des lignes à notre grille
    plt.grid(axis="y", linewidth=0.5)
    plt.show();

    #plt.xticks([1975, 1995, 2015], ['start', 1995, 'end'])
    #plt.yticks([0, 3000])
    #plt.xlim([1975, 2015])
    #plt.ylim([0, 3000])

print(sorted(plt.style.available)) # Liste de tous les styles de matplotlib
```

### Les plots multiples

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    
    plt.subplot(1,2,1)
    plt.plot(years_x, total_y, label="Total", color="orange") 
    plt.title('USA - Total CO2 emissions from electricity production')
    plt.xlabel("Year")
    plt.ylabel("CO2 - M of tons")
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(years_x, coal_y, label="Coal")
    plt.plot(years_x, gas_y, label="Gaz")
    plt.title('USA - CO2 emissions from electricity production')
    plt.xlabel("Year")
    plt.ylabel("CO2 - M of tons")
    plt.legend()
    
    plt.grid(axis="y", linewidth=0.5)
    plt.show();
```

## Avec Pandas

```python
df = pd.DataFrame({ 'coal': coal_y, 'gas': gas_y }, index=years_x)
df
```

```python
x = df.index.values
coal_y = df.coal
gas_y = df.gas

with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    plt.plot(x, coal_y, label="Coal")
    plt.plot(x, gas_y, label="Gaz")
    plt.title('USA - CO2 emissions from electricity production')
    plt.xlabel("Year")
    plt.ylabel("CO2 - M of tons")
    plt.legend()
    plt.grid(axis="y", linewidth=0.5)
    plt.show();
```

## Les différents types de plot

### La représentation des données numériques

#### 1 variable numérique

##### L'histogramme

```python
x = np.random.normal(size=1000)
```

```python
# L'histograme permet une représentation graphique d'une donnée unique et est découpée en 'bins', représentant
# le nombre d'observations par la hauteur de la barre
```

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    plt.hist(x, bins=100)
    plt.grid(axis="y", linewidth=0.5)
    plt.show();
```

##### Le Kde

```python
##### Le 'density plot' peut être vu comme une extension de l'histograme
```

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    sns.histplot(x, bins=100, kde=True);
    plt.grid(axis="y", linewidth=0.5)
    plt.show();
```

#### 2+ variables numériques

##### Le scatterplot

```python
# Le scatterplot permet d'afficher une relation entre 2 variables numériques
```

```python
data = pd.read_csv('https://gist.githubusercontent.com/ssaunier/8044d6a7267223787ed143d0973e3ec6/raw/youtube.csv')
```

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    plt.scatter(data['views'], data['likes'], edgecolor='#333333', alpha=0.75)
    plt.grid(axis="y", linewidth=0.5)
    plt.title('Trending YouTube Videos')
    plt.xlabel('Views')
    plt.ylabel('Likes')
    plt.xscale('log')
    plt.yscale('log')
    plt.show();
```

### La représentation des données catégoriques

#### 1 variable catégorique

##### Le count plot

```python
# Le 'count plot' montre la relation entre une donnée catégorique, une donnée numérique et son occurence
```

```python
tips_df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
```

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    sns.countplot(x="smoker", data=tips_df);
    plt.grid(axis="y", linewidth=0.5)
    plt.show();
```

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    sns.countplot(x="time", hue="smoker", data=tips_df);
    plt.grid(axis="y", linewidth=0.5)
    plt.show();
```

#### 2+ variables catégoriques

##### Le bar plot

```python
# Le 'bar plot' montre la relation entre une donnée catégorique et une donnée numérique
```

```python
fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]

with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    plt.bar(fruits, counts)
    plt.grid(axis="y", linewidth=0.5)
    plt.title('Fruit supply by kind and color')
    plt.xlabel('Fruits')
    plt.ylabel('Count')
    plt.show();
```

##### Le categorical plot

```python
# Le 'categorical plot' donne un résumé de deux variables numérique et catégorique
```

```python
with plt.style.context('seaborn'):
    plt.figure(figsize=(12,3))
    sns.catplot(x='day', y='total_bill', data=tips_df)
    plt.show();
```
